```python
#4)Draw suitable visualization which shows the share of oil reserves by each region in 2010.
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
oil=pd.read_csv('oil.csv')
fig=plt.scatter(oil['Region'],oil['2010'])
plt.xlabel('Regions')
plt.ylabel('2010')
plt.title('share of oil reserves by each region in 2010')
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.show()
```

```python
#1) Which state has produced max milk in year 2013-14?
import numpy as np
import pandas as pd
milk=pd.read_csv('milk.csv')
milk["maxmilk2013_14"] = milk["Cow Milk-2013-14"] + milk["Boffalo Milk-2013-14"] + milk["Goat Milk-2013-14"]
milk[milk["maxmilk2013_14"]==milk["maxmilk2013_14"].max()]["State/ UT Name"]
#milk.head(100)
```

```python
#3)List top 5 milk producing states in each year from 2010 to 2015
milk["y2011"]=milk["Cow Milk-2010-11"] + milk["Boffalo Milk-2010-11"] + milk["Goat Milk-2010-11"]
milk["y2012"]=milk["Cow Milk-2011-12"] + milk["Boffalo Milk-2011-12"] + milk["Goat Milk-2011-12"]
milk["y2014"]=milk["Cow Milk-2013-14"] + milk["Boffalo Milk-2013-14"] + milk["Goat Milk-2013-14"]
milk["y2015"]=milk["Cow Milk-2014-15"] + milk["Boffalo Milk-2014-15"] + milk["Goat Milk-2014-15"]
milk["y2016"]=milk["Cow Milk-2015-16"] + milk["Boffalo Milk-2015-16"] + milk["Goat Milk-2015-16"]


print("Top 5 cities of 2011")
print("\n")
print("\n")
print(milk.sort_values("y2011",ascending=False)["State/ UT Name"].head())
print("Top 5 cities of 2012")
print("\n")
print(milk.sort_values("y2012",ascending=False)["State/ UT Name"].head())
print("Top 5 cities of 2014")
print("\n")
print(milk.sort_values("y2014",ascending=False)["State/ UT Name"].head())
print("Top 5 cities of 2015")
print("\n")
print(milk.sort_values("y2015",ascending=False)["State/ UT Name"].head())
print("Top 5 cities of 2016")
print("\n")
print(milk.sort_values("y2016",ascending=False)["State/ UT Name"].head())
print("\n")
```

```python
#4) List 5 states (if present) whose total milk production is increased for last three years.
milk[milk["y2016"].ge(milk["y2015"]) & milk["y2015"].ge(milk["y2014"])].head()
```

```python
#5) Draw line graph showing total milk production for last 5 years of selected state
y=[11204.0,12088.0,13007.0,9656,10817]
x=[2011,2012,2014,2015,2016]

plt.plot(x,y,linestyle='solid',color='brown')
plt.xlabel('Year')
plt.ylabel('Millk production')
plt.title('Andhra Pradesh')
plt.show()
```

```python
#Draw subplots of pie charts showing % milk productions by cow, buffalo and goat for last 4 years of selected state.
cow = [milk.get_value(0,"Cow Milk-2011-12"),milk.get_value(0,"Cow Milk-2013-14"),milk.get_value(0,"Cow Milk-2014-15"),milk.get_value(0,"Cow Milk-2015-16")]
boffalo = [milk.get_value(0,"Boffalo Milk-2011-12"),milk.get_value(0,"Boffalo Milk-2013-14"),milk.get_value(0,"Boffalo Milk-2014-15"),milk.get_value(0,"Boffalo Milk-2015-16")]
goat =[milk.get_value(0,"Goat Milk-2011-12"),milk.get_value(0,"Goat Milk-2013-14"),milk.get_value(0,"Goat Milk-2014-15"),milk.get_value(0,"Goat Milk-2015-16")]

fig, axes = plt.subplots(2, 2)
plt.subplots_adjust(wspace = 0.5, hspace = 0.3)

axes[0,0].pie(cow,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[0,0].set_title("cow production")

axes[0,1].pie(boffalo,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[0,1].set_title("Boffalo production")

axes[1,0].pie(goat,labels=[2012,2014,2015,2016],autopct='%1.1f%%')
axes[1,0].set_title("Goat production")
plt.show()
print(boffalo)
```
